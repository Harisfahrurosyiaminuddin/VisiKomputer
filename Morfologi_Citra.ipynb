```python
# Thresholding citra agar menjadi citra biner
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```

```python
# Kernel untuk operasi morfologi (misalnya, ukuran 5x5)
kernel = np.ones((5, 5), np.uint8)
```

```python
# Operasi Erosi
erosion = cv2.erode(binary_image, kernel, iterations=1)
```

```python
# Operasi Dilasi
dilation = cv2.dilate(binary_image, kernel, iterations=1)
```

```python
# Operasi Opening (Dilasi -> Erosi)
opening = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel)
```

```python
# Operasi Closing (Erosi -> Dilasi)
closing = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)
```

```python
# Menampilkan hasil operasi morfologi
plt.figure(figsize=(10, 7))
```

```python
plt.subplot(2, 3, 1)
plt.title("Citra Biner")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(2, 3, 2)
plt.title("Erosi")
plt.imshow(erosion, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(2, 3, 3)
plt.title("Dilasi")
plt.imshow(dilation, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(2, 3, 4)
plt.title("Opening")
plt.imshow(opening, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(2, 3, 5)
plt.title("Closing")
plt.imshow(closing, cmap='gray')
plt.axis('off')
```
